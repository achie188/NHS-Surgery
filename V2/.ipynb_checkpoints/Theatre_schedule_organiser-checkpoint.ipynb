```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#IMPORT WAITING LIST
url = r'C:\Users\achie\Github\NHS-Surgery\V2'

url = url + r'\Patient List - Sorted.xlsx'
Clinical_urgency = pd.read_excel(url, sheet_name = 0, header=0, index_col=0, engine='openpyxl')
Clinical_efficiency = pd.read_excel(url, sheet_name = 1, header=0, index_col=0, engine='openpyxl')
Waiting_urgency = pd.read_excel(url, sheet_name = 2, header=0, index_col=0, engine='openpyxl')
Waiting_efficiency = pd.read_excel(url, sheet_name = 3, header=0, index_col=0, engine='openpyxl')
Welfare_urgency = pd.read_excel(url, sheet_name = 4, header=0, index_col=0, engine='openpyxl')
Welfare_efficiency = pd.read_excel(url, sheet_name = 5, header=0, index_col=0, engine='openpyxl')
Number = pd.read_excel(url, sheet_name = 6, header=0, index_col=0, engine='openpyxl')


url = r'C:\Users\achie\Github\NHS-Surgery\V2'

url = url + r'\Input.xlsx'
Theatres = pd.read_excel(url, sheet_name = 1, header=0, index_col=0, engine='openpyxl')

%store -r NDays

start = dt.datetime(2020,12,28)
end = start + dt.timedelta(days=NDays)
```

```python
#GET LIST

%store -r Crit_Type

if Crit_Type == 'CLINICAL URGENCY':
    List_of_data2 = Clinical_urgency.copy()
    
if Crit_Type == 'CLINICAL EFFICIENCY':
    List_of_data2 = Clinical_efficiency.copy()
    
if Crit_Type == 'WAITING URGENCY':
    List_of_data2 = Waiting_urgency.copy()
    
if Crit_Type == 'WAITING EFFICIENCY':
    List_of_data2 = Waiting_efficiency.copy()
    
if Crit_Type == 'WELFARE URGENCY':
    List_of_data2 = Welfare_urgency.copy()
    
if Crit_Type == 'WELFARE EFFICIENCY':
    List_of_data2 = Welfare_efficiency.copy()
    
if Crit_Type == 'VOLUME':
    List_of_data2 = Number.copy()
```

```python
#SPLIT SPECIALTY TYPES

List_of_data = [List_of_data2[List_of_data2['Specialty Descriptions'] == name] 
                    for name in List_of_data2['Specialty Descriptions'].unique()]

Overall_slots = pd.DataFrame()
Overall_sched = pd.DataFrame()
Overall_double = pd.DataFrame()

Total = len(List_of_data)

for type in range(0, Total):
#    type = 2
    Surgery = List_of_data[type].copy()
    
    Scheduled_ops = pd.DataFrame()
    Double_ops = pd.DataFrame()
    
    x = 1

    Surgery['Specialty_rank'] = "None"
    
    while len(Surgery) > 0:
        Op_time = 0
        Surgery = Surgery.reset_index(drop = True)
        for patient_x in range(0, len(Surgery)):
            Op_time = Op_time + Surgery.loc[patient_x, 'New_op_time']

            if Op_time <= 240:
                Surgery.loc[patient_x, 'Specialty_rank'] = x
            else:
                Op_time = Op_time - Surgery.loc[patient_x, 'New_op_time']
            
            if Surgery.loc[patient_x, 'New_op_time'] > 240:
                Surgery.loc[patient_x, 'Specialty_rank'] = "Double slot"
            
        Double_ops = Double_ops.append(Surgery[Surgery['Specialty_rank'] == "Double slot"]) 
        Scheduled_ops = Scheduled_ops.append(Surgery[Surgery['Specialty_rank'] == x])

        Surgery = Surgery[Surgery['Specialty_rank'] == "None"]
        
        x = x + 1 
        
    Clinical_urg = Scheduled_ops.groupby("Specialty_rank")['Hospital Urgency Score'].sum()

    Wait_time = Scheduled_ops.groupby("Specialty_rank")['Waiting_time'].sum()
    
    Welfare_cost = Scheduled_ops.groupby("Specialty_rank")['Welfare_cost'].sum()
    
    No_operations = Scheduled_ops.groupby("Specialty_rank")['Hospital Urgency Score'].count()
    
    Operation_time = Scheduled_ops.groupby("Specialty_rank")['Expected length of operation (mins)'].sum()
    
    Slot_ranking = pd.concat([Clinical_urg, Wait_time, Welfare_cost, No_operations, Operation_time], axis=1)
    
    Type = Scheduled_ops['Specialty Descriptions'].unique()
    Type = Type[0]
    Slot_ranking['Type'] = Type
    
    Slot_ranking.columns = ['Total_urgency', 'Total_wait', 'Total_welfare', 'No_operations', 'Operation_time', 'Specialty']
    
    Slot_ranking = Slot_ranking.round({'Total_wait': 2}) 

    Slot_ranking['Theatre_efficiency'] = Slot_ranking['Operation_time'] / Slot_ranking['No_operations'] 
    
    Overall_slots = Overall_slots.append(Slot_ranking)
    Overall_sched = Overall_sched.append(Scheduled_ops)
    Overall_double = Overall_double.append(Double_ops)
      

if Crit_Type == 'CLINICAL URGENCY':
    Overall_slots = Overall_slots.sort_values(by='Total_urgency', ascending=False)
    
if Crit_Type == 'CLINICAL EFFICIENCY':
    Overall_slots = Overall_slots.sort_values(by='Total_urgency', ascending=False)
    
if Crit_Type == 'WAITING URGENCY':
    Overall_slots = Overall_slots.sort_values(by='Total_wait', ascending=False)
    
if Crit_Type == 'WAITING EFFICIENCY':
    Overall_slots = Overall_slots.sort_values(by='Total_wait', ascending=False)
    
if Crit_Type == 'WELFARE URGENCY':
    Overall_slots = Overall_slots.sort_values(by='Total_welfare', ascending=False)
    
if Crit_Type == 'WELFARE EFFICIENCY':
    Overall_slots = Overall_slots.sort_values(by='Total_welfare', ascending=False)
    
if Crit_Type == 'VOLUME':
    Overall_slots = Overall_slots.sort_values(by='No_operations', ascending=False)


Overall_slots = Overall_slots.reset_index()
Overall_slots['Theatre_rank'] = Overall_slots.index


Overall_sched['Specialty_rank'] = Overall_sched['Specialty_rank'].values.astype(str)
Overall_slots['Specialty_rank'] = Overall_slots['Specialty_rank'].values.astype(str)

Overall_sched['Session_ID'] = Overall_sched['Specialty Descriptions'] + Overall_sched['Specialty_rank']
Overall_slots['Session_ID'] = Overall_slots['Specialty'] + Overall_slots['Specialty_rank']
```

```python
#CREATE SCHEDULE

Schedule = pd.DataFrame()

Type_fill = Overall_slots['Specialty'].unique()
Type_fill = pd.DataFrame(Type_fill)
Type_fill.columns = ['Specialty']
Type_fill.index = Type_fill['Specialty']

Overall_count = 0

Overall_slots['Session'] = "None"

def daterange(start, end):
    for n in range(int((end - start).days)):
        yield start + dt.timedelta(n)       

for date in daterange(start, end):
    weekday = date.strftime('%A')
    day_num = date.weekday() + 1

    for half_day in range(0,2):
        rank_slot = 0
        Updated_slots = Overall_slots[Overall_slots['Session'] == "None"]
        Updated_slots = Updated_slots.reset_index(drop=True)  

        
        slot = (day_num * 2) - 1 + half_day

        theatre_no = 1
        no_slots = 0
        Type_fill['No_slots'] = 0
        
        while theatre_no <= 9:
            if rank_slot + 1 > len(Updated_slots)-1:
                theatre_no = theatre_no + 1
                continue
                
            Type = Updated_slots.loc[rank_slot, 'Specialty']
            Availability = Theatres.loc[Type, slot]
            no_slots = Type_fill.loc[Type, 'No_slots']
                
            if no_slots >= Availability:
                rank_slot = rank_slot + 1
                
            else:
                Schedule.loc[Overall_count, 'Date'] = date
                Schedule.loc[Overall_count, 'Day'] = weekday
                Schedule.loc[Overall_count, 'Session'] = half_day
                Schedule.loc[Overall_count, 'Theatre'] = theatre_no
                Schedule.loc[Overall_count, 'Speciality'] = Type
                Schedule.loc[Overall_count, 'Operations'] = Updated_slots.loc[rank_slot, 'No_operations']
                Schedule.loc[Overall_count, 'Clinical_urgency'] = Updated_slots.loc[rank_slot, 'Total_urgency']
                Schedule.loc[Overall_count, 'Waiting_time'] = Updated_slots.loc[rank_slot, 'Total_wait']
                Schedule.loc[Overall_count, 'Welfare_cost'] = Updated_slots.loc[rank_slot, 'Total_welfare']
                Schedule.loc[Overall_count, 'Operation_time'] = Updated_slots.loc[rank_slot, 'Operation_time']
                Schedule.loc[Overall_count, 'Theatre_effiency'] = Updated_slots.loc[rank_slot, 'Operation_time'] / 240
                Ref = Updated_slots.loc[rank_slot, 'Theatre_rank'] 

                Overall_slots.loc[Ref, 'Date'] = date
                Overall_slots.loc[Ref, 'Day'] = weekday
                Overall_slots.loc[Ref, 'Session'] = half_day

                rank_slot = rank_slot + 1
                theatre_no = theatre_no + 1
                no_slots = no_slots + 1
                Type_fill.loc[Type, 'No_slots'] = no_slots
                
                Overall_count = Overall_count + 1
```

```python
#EFFICIENCY SUMMARY

No_patients = Schedule['Operations'].sum()


Total_clinical = Schedule['Clinical_urgency'].sum()
Total_waiting = Schedule['Waiting_time'].sum()
Total_waiting = round(Total_waiting, 2)
Total_welfare = Schedule['Welfare_cost'].sum()

Avg_clinical = Total_clinical / No_patients
Avg_clinical = round(Avg_clinical, 1)
Avg_wait = Total_waiting / No_patients
Avg_wait = round(Avg_wait, 0)
Avg_welfare = Total_welfare / No_patients
Avg_welfare = round(Avg_welfare, 0)


TNo_patients = len(List_of_data2)
TTotal_clinical = List_of_data2['Hospital Urgency Score'].sum()
TTotal_waiting = List_of_data2['Waiting_time'].sum()
TTotal_welfare = List_of_data2['Welfare_cost'].sum()

TAvg_clinical = List_of_data2['Hospital Urgency Score'].mean()
TAvg_clinical = round(TAvg_clinical,1)
TAvg_waiting = List_of_data2['Waiting_time'].mean()
TAvg_waiting = round(TAvg_waiting,0)
TAvg_welfare = List_of_data2['Welfare_cost'].mean()
TAvg_welfare = round(TAvg_welfare,0)


ANo_patients = TNo_patients - No_patients
ATotal_clinical = TTotal_clinical - Total_clinical
ATotal_waiting = TTotal_waiting - Total_waiting
ATotal_welfare = TTotal_welfare - Total_welfare

AAvg_clinical = ATotal_clinical / ANo_patients
AAvg_clinical = round(AAvg_clinical,1)
AAvg_waiting = ATotal_waiting / ANo_patients
AAvg_waiting = round(AAvg_waiting,0)
AAvg_welfare = ATotal_welfare / ANo_patients
AAvg_welfare = round(AAvg_welfare,0)

Theatre_efficiency = Schedule['Operation_time'].sum() / (len(Schedule)*240)


Efficiency_Summary = ['Patients Scheduled', 'Total Clinical Urgency', 'Average Clinical Urgency',
                      'Total Waiting Time (days)', 'Average waiting time (days)', 'Total Welfare Cost', 'Average Welfare Cost']

Efficiency_Summary = pd.DataFrame(Efficiency_Summary)

Efficiency_Summary.columns = ['Summary']
Efficiency_Summary.index = Efficiency_Summary['Summary']
Efficiency_Summary = Efficiency_Summary.drop('Summary', 1)

#OPERATIONS
Efficiency_Summary.loc['Patients Scheduled', 'Operated_on'] = No_patients
Efficiency_Summary.loc['Total Clinical Urgency', 'Operated_on'] = Total_clinical
Efficiency_Summary.loc['Average Clinical Urgency', 'Operated_on'] = Avg_clinical
Efficiency_Summary.loc['Total Waiting Time (days)', 'Operated_on'] = Total_waiting
Efficiency_Summary.loc['Average waiting time (days)', 'Operated_on'] = Avg_wait
Efficiency_Summary.loc['Total Welfare Cost', 'Operated_on'] = Total_waiting
Efficiency_Summary.loc['Average Welfare Cost', 'Operated_on'] = Avg_wait

#BEFORE
Efficiency_Summary.loc['Patients Scheduled', 'Before'] = TNo_patients
Efficiency_Summary.loc['Total Clinical Urgency', 'Before'] = TTotal_clinical
Efficiency_Summary.loc['Average Clinical Urgency', 'Before'] = TAvg_clinical
Efficiency_Summary.loc['Total Waiting Time (days)', 'Before'] = TTotal_waiting
Efficiency_Summary.loc['Average waiting time (days)', 'Before'] = TAvg_waiting
Efficiency_Summary.loc['Total Welfare Cost', 'Before'] = TTotal_welfare
Efficiency_Summary.loc['Average Welfare Cost', 'Before'] = TAvg_welfare

#AFTER
Efficiency_Summary.loc['Patients Scheduled', 'After'] = ANo_patients
Efficiency_Summary.loc['Total Clinical Urgency', 'After'] = ATotal_clinical
Efficiency_Summary.loc['Average Clinical Urgency', 'After'] = AAvg_clinical
Efficiency_Summary.loc['Total Waiting Time (days)', 'After'] = ATotal_waiting
Efficiency_Summary.loc['Average waiting time (days)', 'After'] = AAvg_waiting
Efficiency_Summary.loc['Total Welfare Cost', 'After'] = ATotal_welfare
Efficiency_Summary.loc['Average Welfare Cost', 'After'] = AAvg_welfare


PDiff = ANo_patients - TNo_patients
CDiff = AAvg_clinical - TAvg_clinical
WDiff = AAvg_waiting - TAvg_waiting
WelDiff = AAvg_welfare - TAvg_welfare

PDiff = round(PDiff,0)
CDiff = round(CDiff,1)
WDiff = round(WDiff,0)
WelDiff = round(WelDiff,0)
```

```python
Patient_day = Overall_slots[['Session_ID', 'Date', 'Day', 'Session']]

Clean_Sched = pd.merge(Overall_sched, Patient_day, how= "left", on=["Session_ID"])

Clean_Sched = Clean_Sched.dropna(subset=['Date', 'Day'])

Clean_Sched = Clean_Sched[['Date', 'Day', 'Session', 'Patient_ID', 'Specialty Code', 'Specialty Descriptions',  
             'Procedure National Description', 'Procedure National Code', 'Date of Admission',
            'Hospital Urgency Score', 'Expected length of operation (mins)', 'Waiting_time', 
            'Needs Ward Bed post operation (0 for no, 1 for yes)',
            ]]
```
