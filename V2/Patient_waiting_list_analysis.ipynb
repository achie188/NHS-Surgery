```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
#IMPORT WAITING LIST
url = r'C:\Users\achie\Github\NHS-Surgery\V2'

url = url + r'\Input.xlsx'
Patient_List = pd.read_excel(url, sheet_name = 0, header=0, index_col=0, engine='openpyxl')
```

```python
#GET NEW WAITING TIME & WELFARE SCORES

today = dt.datetime.today()

Patient_List['Waiting_time'] = (today - Patient_List['Date of Admission']).dt.days

Patient_List['Welfare_cost'] = Patient_List['Waiting_time'] * Patient_List['Hospital Urgency Score']
```

```python
#OVERALL STATISTICS

No_Patients = len(Patient_List)

Tot_Wait = Patient_List['Waiting_time'].sum()
Tot_Urgency = Patient_List['Hospital Urgency Score'].sum()
Tot_Welfare = Patient_List['Welfare_cost'].sum()


Avg_Wait = Patient_List['Waiting_time'].mean()
Avg_Wait = round(Avg_Wait, 2)
Avg_Urgency = Patient_List['Hospital Urgency Score'].mean()
Avg_Urgency = round(Avg_Urgency, 2)
Avg_Welfare = Patient_List['Welfare_cost'].mean()
Avg_Welfare = round(Avg_Welfare, 0)


Max_Wait = Patient_List['Waiting_time'].max()
Max_Urgency = Patient_List['Hospital Urgency Score'].max()
Max_Welfare = Patient_List['Welfare_cost'].max()
```

```python
#SPECIALTY STATISTICS

Waiting_avg = Patient_List.groupby("Specialty Descriptions")["Waiting_time"].mean()
Urgency_avg = Patient_List.groupby("Specialty Descriptions")["Hospital Urgency Score"].mean()
Welfare_avg = Patient_List.groupby("Specialty Descriptions")["Welfare_cost"].mean()
Specialtys_no = Patient_List.groupby("Specialty Descriptions")["Specialty Descriptions"].count()

Specialty_summ = pd.concat([Waiting_avg, Urgency_avg, Welfare_avg, Specialtys_no], axis=1)

Specialty_summ.columns = ['Avg_waiting', 'Avg_urgency', 'Avg_welfare', 'No_patients']

Specialty_summ = Specialty_summ.round({'Avg_waiting': 0, 'Avg_welfare': 1, 'Avg_urgency': 1})
```

```python
#PROCEDURE STATISTICS

Pro_Waiting_avg = Patient_List.groupby(["Specialty Descriptions", "Procedure National Description"])["Waiting_time"].mean()
Pro_Urgency_avg = Patient_List.groupby(["Specialty Descriptions", "Procedure National Description"])["Hospital Urgency Score"].mean()
Pro_Welfare_avg = Patient_List.groupby(["Specialty Descriptions", "Procedure National Description"])["Welfare_cost"].mean()
Pro_Specialtys_no = Patient_List.groupby(["Specialty Descriptions", "Procedure National Description"])["Specialty Descriptions"].count()

Pro_Specialty_summ = pd.concat([Pro_Waiting_avg, Pro_Urgency_avg, Pro_Welfare_avg, Pro_Specialtys_no], axis=1)

Pro_Specialty_summ.columns = ['Avg_pro_waiting', 'Avg_pro_urgency', 'Avg_pro_welfare', 'No_pro_patients']

Pro_Specialty_summ = Pro_Specialty_summ.round({'Avg_pro_waiting': 0, 'Avg_pro_urgency': 1, 'Avg_pro_urgency': 1})
```

```python
#GET TOP VALUES

Top_Specialtys_no = Specialty_summ.nlargest(20,'No_patients')
Top_Specialtys_waiting = Specialty_summ.nlargest(20,'Avg_waiting')
Top_Specialtys_urgency = Specialty_summ.nlargest(20,'Avg_urgency')
Top_Specialtys_welfare = Specialty_summ.nlargest(20,'Avg_welfare')

Top_Procedures_no = Pro_Specialty_summ.nlargest(20,'No_pro_patients')
Top_Procedures_waiting = Pro_Specialty_summ.nlargest(20,'Avg_pro_waiting')
Top_Procedures_urgency = Pro_Specialty_summ.nlargest(20,'Avg_pro_urgency')
Top_Procedures_welfare = Pro_Specialty_summ.nlargest(20,'Avg_pro_welfare')


#Op_time = Patient_List[['Procedure National Description', 'Expected length of operation (mins)']]

#Top_Procedures_no = pd.merge(Top_Procedures_no, Op_time, on='Procedure National Description')
```
